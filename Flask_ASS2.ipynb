````python
Q1. Explain GET and POST methods.
Ans:
    Sure! Here's a brief explanation of the GET and POST methods:

GET method:
- Used to retrieve data from a server.
- Appends data to the URL as query parameters.
- Limited data size (URL length constraints).
- Data is visible in the URL (less secure).
- Caching is possible, making it suitable for non-destructive operations.

POST method:
- Used to send data to a server to create or update a resource.
- Sends data in the request body, not visible in the URL.
- Can handle larger data compared to GET.
- More secure for sensitive information.
- Not cached by default, suitable for destructive operations and sensitive data transmission.
````

````python
Q2. Why is request used in Flask?
Ans:
    In Flask, the "request" object is used to handle incoming HTTP requests from clients (typically web browsers or API clients)
    . It allows Flask applications to access data sent by the client, such as form data, URL parameters, and JSON payloads. The
    "request" object provides an interface to extract and manipulate this incoming data, enabling developers to build dynamic
    and interactive web applications by processing user input and responding accordingly.
````

````python
Q3. Why is redirect() used in Flask?
Ans:
In Flask, the "redirect()" function is used to perform an HTTP redirect response to a different URL. It is used when a route 
needs to redirect the user to another page or route within the application or to an external URL. This helps in implementing 
various features like after a successful form submission, user authentication, or handling page not found scenarios by redire-
cting users to an error page. The "redirect()" function simplifies the process of sending redirection responses and enhances 
the user experience in web applications.
````

````python
Q4. What are templates in Flask? Why is the render_template() function used?
Ans:
    In Flask, templates are files that contain dynamic content and HTML structure. They allow developers to separate the prese-
    ntation layer from the application logic, making it easier to manage and maintain web applications.

The "render_template()" function in Flask is used to render these templates and dynamically inject data into them. It takes the
template file as input and can also accept variables that provide dynamic data to the template. This enables developers to crea-
te dynamic web pages by combining static HTML structure with dynamic content generated by the application. The "render_template
()" function facilitates the process of generating dynamic web pages and improves the maintainability and readability of the 
code by keeping the HTML and Python logic separate.
````
